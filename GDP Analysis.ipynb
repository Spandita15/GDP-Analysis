# GDP Growth of a Country

```python
df_new = df[df["Country Name"] == "India"]
df_new.head()
```

```python
data = [0]
arr = df_new["Value"].values
for i in range(1,len(arr)):
    prev = arr[i-1]
    curr = arr[i]
    gdp_ch = round((curr - prev)*100/prev,2)
    data.append(gdp_ch)
df_new["GDP Growth"] = data
df_new
```

```python
df_new["GDP Growth"].max()
```

# GDP Growth on Whole Dataset

```python
final_data = []
for i in df["Country Name"].unique():
    df_new = df[df["Country Name"] == i]
    data = [0]
    arr = df_new["Value"].values
    for i in range(1,len(arr)):
        prev = arr[i-1]
        curr = arr[i]
        gdp_ch = round((curr - prev)*100/prev,2)
        data.append(gdp_ch)
    df_new["GDP Growth"] = data
    final_data.append(df_new)
df = pd.concat(final_data, axis = 0)
df
```

```python
df.groupby("Country Name")["Value"].mean().sort_values(ascending = False).head(50)
```

# Compare GDP across Countries Advanced

```python
def compare_gdp(lst):
    dfs = []
    for i in lst:
        dfs.append(df[df["Country Code"] == i])
        df_pr = pd.concat(dfs, axis =0)
    fig5 = px.line(df_pr, x = "Year", y = "Value", title = "GDP Comparison - " + "|".join(lst), color = "Country Name") 
    pyo.plot(fig5, filename = "-".join(lst) + ".html", auto_open = False)
compare_gdp(["IND", "CHN", "USA"])
```

# Compare GDP Growth

## GDP GrowthComparison between Specific Countries

```python
def compare_gdp_growth(lst):
    dfs = []
    for i in lst:
        dfs.append(df[df["Country Code"] == i])
        df_pr = pd.concat(dfs, axis =0)
    fig6 = px.line(df_pr, x = "Year", y = "GDP Growth", title = "GDP Growth Comparison - " + "|".join(lst), color = "Country Name") 
    pyo.plot(fig6, filename = "-".join(lst) + ".html", auto_open = False)
compare_gdp_growth(["IND", "CHN", "USA", "BRA"])
```
